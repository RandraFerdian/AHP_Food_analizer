**Menghapus data null**

**extract ke fastfood_clean.csv**


Import librabry


```python
import pandas as pd
import numpy as np
```

Read Data


```python
# Membaca file CSV
df = pd.read_csv('./dataset/fastfood.csv')
display(df.info())
```

seleksi kolom yang diperlukan & menghapus null

1. Kalori (Calories)
2. Lemak Jenuh (Saturated Fat / sat_fat)
3. Natrium (Sodium)
4. Protein


```python
kolom_pilihan = ['restaurant','item','protein' ,'calories', 'total_fat' ,'sodium','cholesterol' ]
df_pil = df[kolom_pilihan]
df_pil.info()

# Menampilkan baris yang masih memiliki nilai null pada df_pil
display(df_pil[df_pil.isnull().any(axis=1)])

#menghapus baris yang memiliki nilai null
df_pil = df_pil.dropna()
df_pil.info()

display(df_pil)
```

ubah ke excel atau csv


```python
#mengexport ke file CSV
df_pil.to_csv('./dataset/fastfood_clean.csv', index=False)
```
